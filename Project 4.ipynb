```python
#import nltk and other libraries
import nltk 
from nltk.corpus import movie_reviews
from nltk import FreqDist
import random
'''
import copy
import csv
from __future__ import division
'''
```

```python
#construct a list of documents, labeled with the appropriate categories - positive or negative
documents = [(list(movie_reviews.words(fileid)), category)
             for category in movie_reviews.categories()
             for fileid in movie_reviews.fileids(category)]
random.shuffle(documents)
```

```python
#get all words in movie_reviews
words = movie_reviews.words()
```

```python
#get the frequency distribution of the words in the movie_reviews corpus
fdist = FreqDist(words)
```

```python
#limit the features, get a list of the most common 2000 words and their frequency
words_2000 = fdist.most_common(2000)
```

```python
#check that this displays the most frequent 2000 words
words_2000
```

```python
#get just the words from the top 2000 words
word_features = [k for (k,v) in words_2000]
```

```python
word_features[0]
```

```python
def document_features(document):
    document_words = set(document)
    features = {}
    for word in word_features:
        features['contains({})'.format(word)] = (word in document_words)
    return features
```

```python
#get the featuresets, split into training set and testing sets and create the classifier
featuresets = [(document_features(d), c) for (d,c) in documents]
train_set, test_set = featuresets[100:], featuresets[:100]
classifier = nltk.NaiveBayesClassifier.train(train_set)
```

```python
#show the accuracy of the classifier
print(nltk.classify.accuracy(classifier, test_set))
```

```python
#get the 30 most informative features
classifier.show_most_informative_features(30)
```

#### Can you explain why these particular features are informative?
These features are the words that appear most often that are associated with either a positive or negative review.  Many of these words are either positive descriptive words like "outstanding", "wonderfully" and "superb" or negative descriptive words like "ridiculous", "awful" and "lame", which will likely indicate a positive or negative review.  Other words refer to popular actors like "damon", "seagal" and "hanks", who tend to be in popular movies or the movies are popular because they are in them. These movies are likely to get more positive reviews.
